# Question 2

```python
input_shape = (150,150,3)
n_filters = 32
kernel = (3,3)
activation_function = 'relu'
max_pooling = (2,2)
inner_layer_size = 64
loss_function = keras.losses.BinaryCrossentropy()
```

### Build the model

```python
inputs = keras.Input(shape=input_shape)
convs  = keras.layers.Conv2D(filters = n_filters, kernel_size = kernel, activation=activation_function, input_shape = input_shape)(inputs)
vectors = keras.layers.MaxPooling2D(pool_size = max_pooling)(convs)
flatten = keras.layers.Flatten()(vectors)
dense = keras.layers.Dense(inner_layer_size, activation='relu')(flatten)
outputs = keras.layers.Dense(1, activation="sigmoid")(dense)
model = keras.Model(inputs, outputs)
```

```python
model.summary()
```

## The convolutionaly layer has 896 parameters

```python
optimizer = keras.optimizers.SGD(learning_rate=0.002, momentum=0.8)
loss = loss_function
model.compile(
        optimizer=optimizer,
        loss=loss,
        metrics=['accuracy']
    )
```

```python
train_gen = ImageDataGenerator(rescale=1./255)
train_ds  = train_gen.flow_from_directory('./data/train',
                                          target_size=(150,150), 
                                          batch_size=20,
                                          shuffle=True,
                                          class_mode = 'binary')
```

```python
val_gen = ImageDataGenerator(rescale=1./255)
val_ds  = val_gen.flow_from_directory('./data/test',
                                          target_size=(150,150), 
                                          batch_size=20,
                                          shuffle=True,
                                          class_mode = 'binary')
```

```python
history = model.fit(train_ds, epochs=10, validation_data=val_ds)
```
